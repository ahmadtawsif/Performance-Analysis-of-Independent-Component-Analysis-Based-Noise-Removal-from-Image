# Both the images are added with random matrix named 'array' which is 100X100

```python
array = np.random.randint(3,size=image1.shape)
im = np.add(image1, array)
plt.imshow(im)
```

# Loading FastICA and PCA

```python
from sklearn.decomposition import FastICA, PCA 

ica =  FastICA(max_iter=1000, tol=0.0001)
```

```python
pca = PCA()
```

```python
imag= ica.fit_transform(im)

H = pca.fit(im).transform(im)
```
